```python
def sig_round_off(number,digits):
    return np.round(number+5*10**(-1*(digits+1)),decimals=digits-1)
```

```python
def Cal_LU(D,g,d):
    A= np.array((D),dtype=float)
    f= np.array((g),dtype=float)
    n = f.size
    for i in range(0,n-1):     # Loop through the columns of the matrix
        for j in range(i+1,n):     # Loop through rows below diagonal for each column
            if A[i,i] == 0:
                print("Error: Zero on diagonal!")
                print("Need algorithm with pivoting")
                break
            m = A[j,i]/A[i,i]
            m = sig_round_off(m,d)
            print("ratio:",m)
            A[j,:] = sig_round_off(A[j,:] - m*A[i,:],d)
            f[j] = sig_round_off(f[j] - m*f[i],d)
    return A,f
```

```python
def Cal_LU_pivot(D,g,d):
    A= np.array((D),dtype=float)
    f= np.array((g),dtype=float)
    n = len(f)
    for i in range(0,n-1):     # Loop through the columns of the matrix
        
        for k in range(i+1,n):
            if np.abs(A[k,i])>np.abs(A[i,i]):
                A[[i,k]]=A[[k,i]]             # Swaps ith and kth rows to each other
                f[[i,k]]=f[[k,i]]
                
        for j in range(i+1,n):     # Loop through rows below diagonal for each column
            m = A[j,i]/A[i,i]
            m = sig_round_off(m,d)
            print("ratio:",m)
            A[j,:] = sig_round_off(A[j,:] - m*A[i,:],d)
            f[j] = sig_round_off(f[j] - m*f[i],d)
    return A,f
```

```python
def Back_Subs(A,f,d):
    n = f.size
    x = np.zeros(n)             # Initialize the solution vector, x, to zero
    x[n-1] = f[n-1]/A[n-1,n-1]    # Solve for last entry first
    for i in range(n-2,-1,-1):      # Loop from the end to the beginning
        sum_ = 0
        for j in range(i+1,n):        # For known x values, sum and move to rhs
            sum_ = sig_round_off(sum_ + A[i,j]*x[j],d)
        x[i] = sig_round_off((f[i] - sum_)/A[i,i],d)
    return x
```

```python
def gauss_elimination_without_pivot(A,f,D):
    print("System of equations:")
    for i in range(A.shape[0]):
        row = ["{}*x{}".format(A[i, j], j + 1) for j in range(A.shape[1])]
        print("[{}] = [{}]".format(" + ".join(row), f[i]))
    B,g = Cal_LU(A,f,D)
    X = Back_Subs(B,g,D)
    return A,f,X
```

```python
def gauss_elimination_with_pivot(A,f,D):
    print("System of equations:")
    for i in range(A.shape[0]):
        row = ["{}*x{}".format(A[i, j], j + 1) for j in range(A.shape[1])]
        print("[{}] = [{}]".format(" + ".join(row), f[i]))
    B,g = Cal_LU_pivot(A,f,D)
    X = Back_Subs(B,g,D)
    return A,f,X
```

### Write a code to perform
### a)  Gauss Jacobi method

```python
def generate_matrix():
    return np.around(np.random.randn(3,3),4)
```

```python
def lu(A):
    
    #Get the number of rows
    n = A.shape[0]
    
    U = A.copy()
    L = np.eye(n, dtype=np.double)
    
    #Loop over rows
    for i in range(n):
            
        #Eliminate entries below i with row operations 
        #on U and reverse the row operations to 
        #manipulate L
        factor = U[i+1:, i] / U[i, i]
        L[i+1:, i] = factor
        U[i+1:] -= factor[:, np.newaxis] * U[i]
        
    return L, U
```

```python
def generate_diagonally_dominant_matrix():
    A = generate_matrix()
    D = np.diag(np.abs(A)) # Find diagonal coefficients
    S = np.sum(np.abs(A), axis=1) - D # Find row sum without diagonal
    L,U = lu(A)
    C = -np.linalg.inv(np.identity(3)+L)*U
    if np.all(D > S) and np.linalg.norm(C,'fro') < 1:
        return A
    else:
        return generate_diagonally_dominant_matrix()
```

### a) Gauss Jacobi Method

```python

ITERATION_LIMIT = 1000

# prints the system
print("System:")
for i in range(A.shape[0]):
    row = ["{}*x{}".format(A[i, j], j + 1) for j in range(A.shape[1])]
    print(" + ".join(row), "=", b[i])
print()

x = np.zeros_like(b)
for it_count in range(ITERATION_LIMIT):
    if it_count != 0:
        print("Iteration {0}: {1}".format(it_count, x))
    x_new = np.zeros_like(x)

    for i in range(A.shape[0]):
        s1 = np.dot(A[i, :i], x[:i])
        s2 = np.dot(A[i, i + 1:], x[i + 1:])
        x_new[i] = (b[i] - s1 - s2) / A[i, i]
        if x_new[i] == x_new[i-1]:
            break

    if np.allclose(x, x_new, atol=0.01, rtol=0.):
        break

    x = x_new

print("Solution:")
print(x)
error = np.dot(A, x) - b
print("Error:")
print(error)
```

```python
def nondd():
    A = generate_matrix()
    D = np.diag(np.abs(A)) # Find diagonal coefficients
    S = np.sum(np.abs(A), axis=1) - D # Find row sum without diagonal
    if all(D < S):
        return A
    else:
        return nondd()
```
